# Trainning monitor

```r
dat = read.table('history.csv',header=TRUE,sep=',')
dat_aug = read.table('history_arg.csv',header=TRUE,sep=',')
head(dat)
```

```r
print(summary(dat$loss))
print(summary(dat_aug$loss))
```

```r
print(summary(dat$accuracy))
print(summary(dat_aug$accuracy))
```

```r
print(summary(dat$val_loss))
print(summary(dat_aug$val_loss))
```

```r
print(summary(dat$val_accuracy))
print(summary(dat_aug$val_accuracy))
```

```r
library(repr)

par(mar=par()$mar + c(0,0,0,3),lwd=2,family='Times')
plot(dat$val_accuracy,type='l',col='black',
     xlab='Epoch',ylab='Accuracy',ylim=c(0,1),
     bty='u',cex.lab=1.5,cex.axis=1.3,
     main='Training of tyCooNN\n(without data augmentation)',
     cex.main=1.7,font.main=2,
     font.axis=2,font.lab=2)
lines(dat$accuracy,col='blue',lty=2)
#lines(3 * dat$val_loss/sum(dat$val_loss),type='l',col='red')
#lines(3 * dat$loss/sum(dat$loss),col='orange',lty=2)
lines(0.3 * dat$val_loss,type='l',col='red')
lines(0.3 * dat$loss,col='orange',lty=2)
axis(side=4,font=2,cex.axis=1.3)
mtext(side=4,line=4,'Loss\nNormalized between 0 and 1',col='red',cex=1.5,font=2)
```

```r
par(mar=par()$mar + c(0,0,0,3),lwd=2,family='Times')
plot(dat_aug$val_accuracy,type='l',col='black',
     xlab='Epoch',ylab='Accuracy',ylim=c(0.9,1),
     bty='u',cex.lab=1.5,cex.axis=1.3,
     main='Training of tyCooNN\n(with data augmentation)',
     cex.main=1.7,font.main=2,
     font.axis=2,font.lab=2)
lines(dat_aug$accuracy,col='blue',lty=2)
lines(0.3 * dat_aug$val_loss + 0.9,type='l',col='red')
lines(0.3 * dat_aug$loss + 0.9,col='orange',lty=2)
axis(side=4,font=2,cex.axis=1.3)
mtext(side=4,line=4,'Loss\nNormalized between 0 and 1',col='red',cex=1.5,font=2)
```

```r
df = rbind(dat,dat_aug)
nrow(df)
```

```r
library(repr)
options(repr.plot.width=10, repr.plot.height=7)
par(mar=par()$mar + c(0,0,0,12),lwd=2,family='Times')
plot(df$val_accuracy,type='l',col='black',lwd=2,
     xlab='Epoch',ylab='Accuracy',ylim=c(0,1),
     bty='u',cex.lab=1.5,cex.axis=1.3,
     main='Training of tyCooNN (IVT)',
     cex.main=1.7,font.main=2,
     font.axis=2,font.lab=2)
lines(df$accuracy,col='blue',lty=2,lwd=2)
lines(0.3 * df$val_loss,type='l',col='red',lwd=2)
lines(0.3 * df$loss,col='orange',lty=2,lwd=2)
axis(side=4,font=2,cex.axis=1.3)
mtext(side=4,line=4,'0.3 * Loss',col='red',cex=1.5,font=2)
abline(v=100,col='magenta',lwd=4)
rect(0,0,100,1,col=rgb(230/256,230/256,230/256,alpha=0.5),border=NA)
rect(100,0,120,1,col=rgb(230/256,0,230/256,alpha=0.1),border=NA)
par(xpd=TRUE)
legend(x=c(0,90),y=c(1.11,1.11),
       legend=c('without augmentation','with augmentation'),horiz=TRUE,bty='n',
       fill=c(rgb(230/256,230/256,230/256,alpha=0.5),rgb(230/256,0,230/256,alpha=0.1)),
       border=NA)
legend(125,1,legend=c('val_accuracy','accuracy','val_loss','loss'),bty='n',
      col=c('black','blue','red','orange'),lty=c(1,2,1,2))
```
